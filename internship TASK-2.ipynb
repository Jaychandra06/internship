```python
def area(r):
    return 3.14 * r * r

r = float(input("Enter radius: "))
print("Area =", area(r))
```

```python
def strlen(s):
    c = 0
    for i in s:
        c += 1
    return c

s = input("Enter string: ")
print("Length =", strlen(s))
```

```python
def large(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= c:
        return b
    else:
        return c

a = int(input())
b = int(input())
c = int(input())

print("Largest =", large(a, b, c))
```

```python
def prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

n = int(input("Enter number: "))

if prime(n):
    print("Prime")
else:
    print("Not Prime")
```

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n - 1) + fib(n - 2)

n = int(input("Enter n: "))
print("Fibonacci =", fib(n))
```

```python
def rev(s):
    if s == "":
        return s
    return rev(s[1:]) + s[0]

s = input("Enter string: ")
print("Reverse =", rev(s))
```

```python
n = int(input("Enter n: "))

s = 0

for i in range(1, n + 1):
    s += i

print("Sum =", s)
```

```python
a = list(map(int, input("Enter numbers: ").split()))

c = 0

for i in a:
    if i % 2 == 0:
        c += 1

print("Even count =", c)
```
